# Extracción del temario de Intervención Sociocomunitaria 
### El propósito de este proyecto es descargar automáticamente el temario para las oposiciones de Intervención Sociocomunitaria de un sitio web que limita la descarga directa del contenido.

```python
import re
import os
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
```

## Extracción de los link del temario

En cada página, el código itera sobre el temario para extraer los enlaces de los temas.

```python
url = 'https://www.oposinet.com/temarios/temario-intevencion-sociocomunitaria/temario-3-intervencion-sociocomunitaria/page/4/'

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

links = soup.find_all('a', itemprop='url', href=True)

hrefs = [link['href'] for link in links]

print("Enlaces extraídos:")
for href in hrefs:
    print(href)

def access_link(href):
    if not href.startswith('http'):
        href = requests.compat.urljoin(url, href)
    
for href in hrefs:
    access_link(href)
```

## Procesamiento y almacenamiento del temario
Esta última parte del código itera sobre los enlaces, procesa el texto, lo limpia y lo guarda en archivos .txt.

```python
def extraer_texto_limpio(url):
    response = requests.get(url)
    
    if response.status_code == 200:
        html_content = response.text
        
        soup = BeautifulSoup(html_content, 'html.parser')
        
        texto_limpio = soup.get_text(separator='\n', strip=True)
        return texto_limpio
    else:
        print(f"Error al acceder a la página web: {response.status_code}")
        return None



for i, enlace in enumerate(enlaces):
    texto_completo = extraer_texto_limpio(enlace)

    if texto_completo:
        nombre_archivo = f"pagina_web_{i}.txt"
        with open(nombre_archivo, 'w', encoding='utf-8') as file:
            file.write(texto_completo)
        
        print(f"Texto extraído y guardado en '{nombre_archivo}' para el enlace: {enlace}")
    else:
        print(f"No se pudo extraer el texto para el enlace: {enlace}")
```
